```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def checkout(self):
        if self.is_available:
            self.is_available = False
            return True
        else:
            return False

    def return_book(self):
        self.is_available = True


class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.checked_out_books = []

    def checkout_book(self, book):
        if book.checkout():
            self.checked_out_books.append(book)
            return True
        else:
            return False

    def return_book(self, book):
        if book in self.checked_out_books:
            book.return_book()
            self.checked_out_books.remove(book)
            return True
        else:
            return False


class Library:
    def __init__(self):
        self.available_books = []

    def add_book(self, book):
        self.available_books.append(book)

    def display_inventory(self):
        print("Library Inventory:")
        for book in self.available_books:
            print(f"{book.title} by {book.author} - {'Available' if book.is_available else 'Not Available'}")



if __name__ == "__main__":

    library = Library()


    num_books = int(input("Enter the number of books to add to the library: "))
    for _ in range(num_books):
        title = input("Enter book title: ")
        author = input("Enter book author: ")
        new_book = Book(title, author)
        library.add_book(new_book)


    library.display_inventory()


    member_name = input("Enter member name: ")
    member_id = int(input("Enter member ID: "))
    member = Member(member_name, member_id)


    checkout_title = input("Enter the title of the book to check out: ")
    checkout_author = input("Enter the author of the book to check out: ")
    checkout_book = next((book for book in library.available_books if book.title == checkout_title and book.author == checkout_author), None)

    if checkout_book:
        if member.checkout_book(checkout_book):
            print(f"{member.name} checked out {checkout_book.title} by {checkout_book.author}")
        else:
            print(f"{checkout_book.title} by {checkout_book.author} is not available.")
    else:
        print("Book not found in the library.")


    library.display_inventory()


    print(f"{member.name}'s Checked-out Books:")
    for book in member.checked_out_books:
        print(f"{book.title} by {book.author}")


    return_title = input("Enter the title of the book to return: ")
    return_author = input("Enter the author of the book to return: ")
    return_book = next((book for book in member.checked_out_books if book.title == return_title and book.author == return_author), None)

    if return_book:
        member.return_book(return_book)
        print(f"{member.name} returned {return_book.title} by {return_book.author}")
    else:
        print("Book not found in the member's checked-out list.")

    library.display_inventory()

    print(f"{member.name}'s Checked-out Books:")
    for book in member.checked_out_books:
        print(f"{book.title} by {book.author}")
```
